```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.feature_extraction.text import CountVectorizer

# membaca dataset dari file excel
df = pd.read_excel('sample_data/Dataset Resep Makanan.xlsx')

# membuat model KNN dengan euclidean distance
k = 3
knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')

# membuat vektor untuk resep makanan
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['resep makanan'])

# mengambil label energi dari dataset
y = df['energi']

# melatih model KNN
knn.fit(X, y)

# fungsi untuk memprediksi energi makanan berdasarkan input
def predict_energy(input_text, knn, vectorizer):
    if input_text in df['resep makanan'].values:
          energy = df.loc[df['resep makanan']==input_text, 'energi'].iloc[0]
    else:
      # mengubah input menjadi vektor
      input_vec = vectorizer.transform([input_text])
      # mencari K tetangga terdekat
      neighbors = knn.kneighbors(input_vec, return_distance=False)
      # mengambil label energi dari K tetangga terdekat
      labels = y.iloc[neighbors[0]]
      # menghitung rata-rata energi dari K tetangga terdekat
      energy = labels.mean()
    return energy

# contoh penggunaan fungsi predict_energy
input_text = input("Masukkan Nama Resep Makanan: ")
input_text.lower()
energy = predict_energy(input_text, knn, vectorizer)
print(f"Resep makanan '{input_text}' diperkirakan memiliki energi sebesar {energy}")
if energy >= 0 and energy <= 200  :
    print('Makanan dibolehkan untuk Diet')
elif energy >200 and energy <=300 :
    print('Makanan diijinkan untuk Diet')
elif energy >300 :
    print('Makanan dilarang untuk diet')
```
